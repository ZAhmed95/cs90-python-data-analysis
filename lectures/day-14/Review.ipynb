## Step 1: Hypotheses Generation

No code involved here, just thinking. You've got a problem you want the answer to, so you generate a bunch of hypotheses to determine what kind of features most likely impact the result you want.

Remember that the hypotheses you come up with might not exactly match the data you find, i.e. there may be features in your hypotheses that's not in th data, or there may be features in the data that you didn't think of in your hypotheses.

In the first situation, you can either: try to get more data that contains your hypotheses features, or simply abandon that hypotheses. You can't test something that you don't have data for.

In the second situation, you can't really complain; you got even more data than you wanted. No data scientist complains about too much data. If you think the new features in the data will be significant to the outcome, then create new hypotheses involving them. However, there may be some data that's completely irrelevant to your question, in that case just ignore it.

## Step 2: Data Exploration

```python
# One of the simplest starting steps to do with pandas is use the dataframe 'describe' function
# for numeric data, it outputs things like mean, std, percentiles, etc.
# for categorical data, it outputs unique count, mode, etc.
desc = features.describe()
desc
```

```python
# create a filter that has 'True' in cells that are outliers
outlier_filter = features.apply(lambda x: np.abs(x - desc.loc['mean', x.name]) > 3 * desc.loc['std', x.name])
outlier_filter
```

```python
# count how many outliers are in each column
outlier_count = outlier_filter.apply(np.sum)
outlier_count
```

```python
# there's another important step usually done during data exploration that we didn't talk about.
# That is, finding which of the features are the best to choose for our model.
# We can use panda's 'corr' method (stands for correlation) to see how the features correlate to each other

# first, create a copy of the features df and add the target column
data = features.copy()
data['MEDV'] = target
data
```

```python
# now call the corr method
# this will give us the correlation score for each feature in relation to every other feature
# we round the values to 2 decimal places for convenience
corr = data.corr().round(2)
```

```python
# It's kind of hard to quickly spot which features are closely related with just a table like this
# better to use a heatmap, where the values will become color coded
plt.imshow(corr, cmap='seismic')
columns = data.columns
plt.yticks(np.arange(len(columns)), columns)
plt.xticks(np.arange(len(columns)), columns, rotation='vertical')
plt.show()
```

This heatmap is a much easier way to see where the high correlation values are, because the human brain naturally understands color differences better than trying to analyze a 2D matrix of numbers.

In this heatmap, the darker red colors mean a correlation value closer to 1 (highly positive), and the darker blue values mean a correlation closer to -1 (highly negative). The closer the color is to white, the closer the correlation is to zero (no correlation). So, we are looking for the darker colors (in either direction). Note that the main diagonal from top left to bottom right, where each feature correlates with itself, is all dark red (r = 1.0). This is obvious, because of course the correlation between an array of data and itself would be 1, because they match perfectly.

From the heatmap, we can judge 2 things:

1. RM has a high positive relation to MEDV, and LSTAT has a high negative relation. These would be good candidates for regression.
2. TAX and RAD have high correlation, so we shouldn't use these features together in the same model. If two features are highly correlated, then it becomes hard to test them independently, because if one changes, the other usually changes as well.

## Data Cleaning

## Feature Engineering

```python
# of the two features we chose, RM and LSTAT, make plots of them against MEDV to see what it looks like
# RM vs MEDV
plt.plot(features['RM'], target, '.')
plt.show()
```

```python
# LSTAT vs MEDV
plt.plot(features['LSTAT'], target, '.')
plt.show()
```

from the plots above, we can clearly see the positive correlation between RM and MEDV, and the negative correlation between LSTAT and MEDV.

Note, however, that while the relationship between RM and MEDV looks pretty linear, the relationship between LSTAT and MEDV definitely doesn't look linear.

In this kind of situation, feature engineering comes in handy: we will transform LSTAT in a way that will give us a linear relationship.

```python
features['LSTAT_2'] = np.log(features['LSTAT'])
plt.plot(features['LSTAT_2'], target, '.')
plt.show()
```

```python
# as you can see, this data looks a lot more linear
```

## Model Creation

```python
# now, to train our model
# If you have a single dataset, you usually don't want to put ALL the data into the model
# because then, you'd have no data left to test on.
# usually, you take 20% of the data as 'test' data, and the remaining 80% as 'training' data
# the training data is what you use to fit the model, and then you test on the test data to see how well it does

# It's not very hard to manually split the data into an 80-20 split,
# but why bother when sklearn has built in functions that do it for you
from sklearn.model_selection import train_test_split
```

```python
# choose only the 2 features we will use as our inputs
X = features[['RM', 'LSTAT_2']]
# name the target dataframe Y, cuz why not
Y = target

# now split the data into training and testing sets
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2)
print(X_train.shape, X_test.shape, Y_train.shape, Y_test.shape)
```

```python
# train the model
from sklearn import linear_model
lr = linear_model.LinearRegression()
lr.fit(X_train, Y_train)
```

```python
# let's see how it does
Y_predicted = lr.predict(X_test)
```

```python
# score the model
lr.score(X_test, Y_test)
```

not too bad. Like I've said before, in analyzing real life data, you are very unlikely to see crazy r2 values above 0.9, unless you have an extremely good model or the data relationship is very obvious to see. This kind of r2 value is what you would consider to be pretty good, and then you would attempt to make it even higher, by doing stuff like adding more features, choosing a better model, getting more data to train on, etc.
