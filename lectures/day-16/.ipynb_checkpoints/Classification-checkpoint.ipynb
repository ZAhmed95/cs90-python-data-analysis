```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import classification_report
```

```python
# load the digits dataset from sklearn.
# this dataset contains ~1700 8x8 greyscale images of the digits 0-9
# each 'pixel' of the images is a value from 0-16, representing greyscale color values
dataset = datasets.load_digits()
```

```python
images = dataset.images
data = dataset.data
target = dataset.target
```

```python
print(images.shape, data.shape, target.shape)
```

```python
# take a look at what some of the digit images look like
# as you can see, 8x8 images are extremely low resolution
for i in range(10):
    plt.imshow(images[i], cmap="Greys")
    plt.show()
```

```python
# split the data into train/test sets, like we did for our previous datasets
train_X, test_X, train_Y, test_Y = train_test_split(data, target, test_size=0.2)
```

```python
print(train_X.shape, test_X.shape, train_Y.shape, test_Y.shape)
```

```python
# here's where things get different: instead of using LinearRegression from sklearn.linear_model,
# we use SVC (Support Vector Classifier) from sklearn.svm
# LinearRegression tries to output a single value which can be any real number,
# while SVC tries to output a single choice from a limited set of 'classes'
# This is the main difference between regression and classification
classifier = SVC(gamma=0.001, kernel='linear')
```

```python
# The way to model the data is still the same: call the fit() method
classifier.fit(train_X, train_Y)
```

```python
# same here, call the predict method to see how the classifier does
predicted = classifier.predict(test_X)
```

```python
print(classification_report(test_Y, predicted))
```

```python
classifier.score(test_X, test_Y)
```

```python
# Let's test it ourselves, by creating our own 8x8 digit:
digit = np.array([
    [0, 9, 9, 9, 9, 9, 9, 0],
    [0, 0, 0, 0, 0, 9, 0, 0],
    [0, 0, 0, 0, 9, 9, 0, 0],
    [0, 0, 0, 9, 9, 0, 0, 0],
    [0, 0, 0, 9, 0, 0, 0, 0],
    [0, 0, 9, 9, 0, 0, 0, 0],
    [0, 0, 9, 0, 0, 0, 0, 0],
    [0, 9, 0, 0, 0, 0, 0, 0],
])
plt.imshow(digit, cmap="Greys")
plt.show()
```

```python
# remember, the model can't take an 8x8 2D image array, we have to flatten it into a 1D array
result = classifier.predict(digit.reshape(1,64))
print(result[0])
```
